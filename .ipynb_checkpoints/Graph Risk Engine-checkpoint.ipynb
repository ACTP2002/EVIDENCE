```python
def nkey(t, _id):
    return f"{t}:{_id}"
```

```python
def update_event_into_graph(event, graph):

    user = f"user:{event['user_id']}"
    ip = f"ip:{event['ip']}"
    device = f"device:{event['device_id']}"


    # Add nodes
    graph.add_node(user, type="user")
    graph.add_node(ip, type="ip")
    graph.add_node(device, type="device")

    # Edge attributes (store entire event for simplicity)
    edge_data = {
        "event_id": event.get("event_id"),
        "event_time": event.get("event_time"),
        "stream_type": event.get("stream_type"),
        "event_type": event.get("event_type"),
        "data": event.get("data")
    }

    # Event as edges
    graph.add_edge(user, ip, relation="user_ip_event", **edge_data)
    graph.add_edge(user, device, relation="user_device_event", **edge_data)
    
    if event['event_type'] == 'buy' or event['event_type'] == 'sell':
        stock = f"stock:{event['data']['stock_id']}"
        graph.add_node(stock, type='stock')
        graph.add_edge(user, stock, relation='user_stock_relation', **edge_data)

    return graph


def process_transfer_event(event, graph, txn_cache):
    txn_id = event.get("txn_id")
    if not txn_id:
        return

    # store event by txn_id
    bucket = txn_cache.setdefault(txn_id, {})
    bucket[event["event_type"]] = event

    # if both sides present â†’ create flow edge
    if "txn_out" in bucket and "txn_in" in bucket:
        out_ev = bucket["txn_out"]
        in_ev = bucket["txn_in"]

        sender = f"user:{out_ev['user_id']}"
        receiver = f"user:{in_ev['user_id']}"

        graph.add_node(sender, type="user")
        graph.add_node(receiver, type="user")

        graph.add_edge(
            sender,
            receiver,
            relation="fund_flow",
            txn_id=txn_id,
            amount=out_ev["data"]["amount"],
            currency=out_ev["data"]["currency"],
            event_time=out_ev["event_time"]
        )

        # remove after pairing
        txn_cache.pop(txn_id)
    
        
```

```python
def summarize_graph(graph: nx.MultiDiGraph):
    print("\n=== GLOBAL GRAPH SUMMARY ===")
    print("nodes:", graph.number_of_nodes())
    print("edges:", graph.number_of_edges())

    # print nodes grouped by type
    users = [n for n, a in graph.nodes(data=True) if a.get("type") == "user"]
    ips = [n for n, a in graph.nodes(data=True) if a.get("type") == "ip"]
    devices = [n for n, a in graph.nodes(data=True) if a.get("type") == "device"]
    print(f"users={len(users)}, ips={len(ips)}, devices={len(devices)}")

    # show neighbors for a few users
    print("\n--- USER NEIGHBORS (sample) ---")
    for u in sorted(users)[:5]:
        neigh = sorted(set(graph.successors(u)))
        print(u, "->", neigh)

    # show shared IPs (IP connected to multiple users)
    print("\n--- SHARED IPs ---")
    for ip in sorted(ips):
        connected_users = set()
        for src, _, _k, _attrs in graph.in_edges(ip, keys=True, data=True):
            if src.startswith("user:"):
                connected_users.add(src)
        if len(connected_users) >= 2:
            print(ip, "connected_users:", sorted(connected_users))

    # show shared devices (device connected to multiple users)
    print("\n--- SHARED DEVICES ---")
    for d in sorted(devices):
        connected_users = set()
        for src, _, _k, _attrs in graph.in_edges(d, keys=True, data=True):
            if src.startswith("user:"):
                connected_users.add(src)
        if len(connected_users) >= 2:
            print(d, "connected_users:", sorted(connected_users))

    # print a few edges (with key)
    print("\n--- SAMPLE EDGES (first 12) ---")
    count = 0
    for u, v, k, attrs in graph.edges(keys=True, data=True):
        print(f"{u} -> {v} (key={k}) | {attrs.get('stream_type')}:{attrs.get('event_type')} | event_id={attrs.get('event_id')}")
        count += 1
        if count >= 12:
            break
            
def visualize_graph(graph, title="Graph View"):
    plt.figure(figsize=(12, 8))

    # Layout algorithm (spring = force-directed layout)
    pos = nx.spring_layout(graph, k=0.5, seed=42)

    # Separate node colors by type
    node_colors = []
    for node, attrs in graph.nodes(data=True):
        if attrs.get("type") == "user":
            node_colors.append("red")
        elif attrs.get("type") == "ip":
            node_colors.append("blue")
        elif attrs.get("type") == "device":
            node_colors.append("green")
        else:
            node_colors.append("gray")

    # Draw nodes
    nx.draw_networkx_nodes(graph, pos, node_color=node_colors, node_size=800)

    # Draw edges
    nx.draw_networkx_edges(graph, pos, alpha=0.6)

    # Draw labels
    nx.draw_networkx_labels(graph, pos, font_size=8)

    plt.title(title)
    plt.axis("off")
    plt.show()
    





# Build the global graph
G = nx.MultiDiGraph()
txn_cache = {}

for ev in EVENTS:
    update_event_into_graph(ev, G)

    if ev["event_type"] in ("txn_in", "txn_out"):
        process_transfer_event(ev, G, txn_cache)
        
        
CG = nx.MultiDiGraph()
cg_txn_cache = {}

for ev in EVENTS_CASE:
    update_event_into_graph(ev, CG)

    if ev["event_type"] in ("txn_in", "txn_out"):
        process_transfer_event(ev, CG, cg_txn_cache)

summarize_graph(G)
visualize_graph(G, title="Fraud Graph - Global View")

summarize_graph(CG)
visualize_graph(CG, title="Fraud Graph - Case View")


```

```python

def graph_to_json_dict(G):
    nodes = []
    for n, attrs in G.nodes(data=True):
        nodes.append({
            "id": n,
            **attrs
        })

    edges = []
    for u, v, k, attrs in G.edges(keys=True, data=True):
        # make sure everything is JSON-serializable
        clean = {}
        for kk, vv in attrs.items():
            # if you stored datetime objects, convert to str
            if hasattr(vv, "isoformat"):
                clean[kk] = vv.isoformat()
            else:
                clean[kk] = vv
        edges.append({
            "source": u,
            "target": v,
            "key": str(k),
            **clean
        })

    return {
        "directed": G.is_directed(),
        "multigraph": True,
        "node_count": G.number_of_nodes(),
        "edge_count": G.number_of_edges(),
        "nodes": nodes,
        "edges": edges
    }

def export_graph_json(G, path="graph.json"):
    payload = graph_to_json_dict(G)
    with open(path, "w", encoding="utf-8") as f:
        json.dump(payload, f, ensure_ascii=False, indent=2)
    return path
```

```python
export_graph_json(G, "global_graph.json")
export_graph_json(CG, "case_graph.json")
```
